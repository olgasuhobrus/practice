```


def pascal_triangle(n): 
    for row in range(0, n): 
        for x in range(0, row + 1):
            a = 1
            if (x > row - x): 
                x = row - x 
            for i in range(0 , x): 
                a = a * (row - i) 
                a = a // (i + 1)
            print(a, " ", end = "") 
        print()
        

while True:
    try:
        n = int(input('Enter power'))
        if n < 0:
            raise ValueError
        break
    except ValueError:
        print("Enter valid number.") 


pascal_triangle(n+1) 
```

```

import math
while True:
    try:
        volume = int(input('Enter desired volume of one gathering (16, 24 or 32 pages): '))
        if (volume != 16) and (volume != 24) and (volume !=32):
            raise ValueError
        break
    except ValueError:
        print('Please, enter valid number.') 
    
while True:
    try:
        pages = int(input('Enter number of pages (Must not exceed 1280. May be rounded up to fit the number of gatherings): '))
        if (pages < 0) or (pages > 1280):
            raise ValueError
        if (pages % volume) != 0:
            pages = int(math.ceil(pages / volume)) * volume
            if pages> 1280:
                print('You may want to change number of pages in a gathering')
                raise ValueError 
        break
    except ValueError:
        print('Number of pages must be between 1 - 1280.')

print(f'Printed book will contain {pages} pages')

while True:
    try:
        parameter = bool(int(input('Would you like to see a complete list of pages in each gathering? Yes - enter 1, No - enter 0: ')))
        break
    except ValueError:
        print('Please, enter 1 or 0.')  

@decorator(pages, volume, parameter)
def sum_of_sheets(pages, volume):
    gatherings_total = int(pages/volume)
    print(f'Book contains {gatherings_total} gatherings.')
    sum_of = []
    print(f'Pages in each gathering:')
    for i in range(gatherings_total):
        gathering = []

        for x in range(int(volume/4)):
            gathering.append(volume - (x * 2) + volume * i)
            gathering.append((x + 1) + x  + volume*i)
            gathering.append((x + 2) + x  + volume*i)
            gathering.append(volume - 1 - (x * 2)  + volume*i)
        
        print(*gathering)
        sum_of.append(gathering)
    return sum_of


def decorator(pages, volume, parameter):
    def wrap(func):
        if parameter:
            sum_of = func(pages, volume)            
            print('\n', sum_of, end='')
        else:
            func(pages, volume)
    return wrap


```
